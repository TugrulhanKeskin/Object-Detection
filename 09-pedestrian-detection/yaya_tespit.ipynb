```python
import cv2
import os 
import matplotlib.pyplot as plt
```

```python
file = os.listdir()
img_path_list = []

for img in file :
    if img.endswith(".jpg"):
        img_path_list.append(img)
print(img_path_list)
```

```python
# Hog tanımlayıcısı 
hog = cv2.HOGDescriptor()
# HOG tanımlayıcısına SVM sınıflandırıcısını yükleme
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

for img_path in img_path_list:
    print(img_path)
    
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # HOG tanımlayıcısını kullanarak insanları tespit etme
    (rects, weights) = hog.detectMultiScale(img, padding=(8,8), scale=1.05)
    # Tespit edilen insanları dikdörtgen içine alıp resme çizdirme
    for (x, y, w, h) in rects:
        cv2.rectangle(img, (x, y), (x + w, y + h), (199, 0, 0), 2)
    
    plt.figure(), plt.imshow(img), plt.axis("off"), plt.show()
```
