```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read image
img = cv2.imread("london.jpg", 0)
plt.figure(), plt.imshow(img, cmap="gray"), plt.axis("off")
```

```python
edges = cv2.Canny(image = img, threshold1 = 0, threshold2 = 255)
plt.figure(), plt.imshow(edges, cmap="gray"), plt.axis("off")
```

suda kenar olarak algılanan çok fazla detay var bu detaylardan kurtulmak için resime özgü threshold değerleri belirliyoruz


```python
med_val = np.median(img)
print(med_val)

low = int(max(0, (1 - 0.33)*med_val))
high = int(min(255, (1 + 0.33)*med_val))

print(low)
print(high)
```

```python
edges = cv2.Canny(image = img, threshold1 = low, threshold2 = high)
plt.figure(), plt.imshow(edges, cmap="gray"), plt.axis("off")
```

```python
#blur ekleme işlemi
# k size arttıkça bulanıklık artar dolayısıyla gereksiz kenarlardan kurtuluruz   
blurred_img = cv2.blur(img, ksize=(7,7))
plt.figure(), plt.imshow(blurred_img, cmap="gray"), plt.axis("off")
```

```python
#blur image ın low ve high değerlerini bulma
med_val = np.median(blurred_img)
print(med_val)

low = int(max(0, (1 - 0.33)*med_val))
high = int(min(255, (1 + 0.33)*med_val))   

print(low)
print(high)
```

```python
#blur uygulanmış hali ile gereksiz kenarları azaltma
edges = cv2.Canny(image = blurred_img, threshold1 = low, threshold2 = high)
plt.figure(), plt.imshow(edges, cmap="gray"), plt.axis("off")
```
